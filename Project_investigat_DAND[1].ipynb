## detecting outliers

```python
sns.boxplot(df['Age']);
```

```python
## what is the mean age? 
df['Age'].describe()
```

```python
#remove the outlier

new_df=df[(df.Age < 80 ) & (df.Age >= 0 )]

new_df['Age'].describe()
```

```python
sns.boxplot(new_df['Age']);
```

```python
bin_edges = [0,13,40,90]
```

```python

bin_names = ['child', 'adult', 'seniors']
```

```python
# Creates age rating column
new_df['age_rate'] = pd.cut(new_df['Age'], bin_edges, labels=bin_names)

# Checks for successful creation of this column
new_df.head()
```

# relation plot between two var. 

##  from the below we can see that seniors attended more than others and at same time no scholarship granted to them . that would open our eyes to different angel of the data , seniors are able to pay their medical requirments.
```python
sns.factorplot(x='Scholarship', y='attendnace', hue='age_rate', data=new_df, kind='bar').set(title='Scholarship affection on attendnace');
```
